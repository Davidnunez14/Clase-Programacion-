1. Cree una imagen de un canal, es decir con una forma (alto, ancho) del tamaño que desee, donde la diagonal principal hacia arriba sea blanca y desde la diagonal hacia abajo sea negra.


```python
import numpy as mm # Se usa la biblioteca numpy para generar los datos necesarios y se le asigna un nombre a los mismos en este caso mm.
from skimage import data# Se usa la biblioteca skimage para ejecutar junto a numpy el procesamiento de imagen.
import matplotlib.pyplot as plt # Finalmente se usa la blioteca matplotlib para generar el grafico que produciran las bibliotecas (numpy y skimage)
Ancho=250# Se le da un valor x a ancho que sera el rango a medir en la pareja ordenada (ejes) de (x,y) en un plano cartesiano.
imagen=[]# Se genera la defincion de una lista vacia para el acceso a la lectura.
pixel_Blanco=[250,250,250]# Se le asigan el numero al pixel a seleccionar en este caso 250 hace referencia al color blanco.
pixel_Negro=[0,0,0]# Expuesto lo anterior es el mismo procedimiento en este caso 0 hace refencia al 
for m in range (0,Ancho):# Se define un rango inicial de 0 al tamaño x que se le dara a la variable (Ancho)
  fila=[]# Se genera la defincion de una lista vacia para el acceso a la lectura en este caso con el nombre fila.
  for l in range (0,Ancho):# Se define un rango inicial de 0 al tamaño x que se le dara a la variable (Ancho)
    if m>=l :# En este caso el if evalua y si el resultado es positivo (True) ejecuta el codigo y sino (False) y no lo ejecutaria.
      fila.append(pixel_Negro)# Usamos en append para agregar un elemento al final de la lista fila y en el componente pixel_Negro.
    else:# Posterior la evalaucion del if el else dice si la condicion se cumple ejecuta y sino no.
      fila.append(pixel_Blanco) #Usamos en append para agregar un elemento al final de la lista fila y en el componente pixel_Blanco.
  imagen.append(fila)# Usamos en append para agregar un elemento al final de la lista imagen y en el componente de lista fila.
imagen_Negra=mm.array(imagen)# El array administra los datos fijos y genera la secuencia del codigo.
plt.imshow(imagen_Negra)# El imshow muestra los datos de las variables como una imagen junto a la biblioteca matplotlib.
print(imagen_Negra)# Finalmente se imprime el resultado grafico.
```

**IMAGEN  FINAL APLICANDO EL METODO DE RESTAURACION:**


---


**NOTA:** EJECUTAR EL CODIGO DEPUES DE HABER EJECUTADO LOS ANTERIORES DE LA PARTE SUPERIOR:

```python
imagen= data.rocket()# Desde skimage se importa x imagen usando un data.
plt.imshow(imagen)#Usamos el plt.imnshow para mostrar los datos como una imagen 2d. 
plt.show# El plt.show nos permite imprimir la imagen asiganada en el data.
img = rgb2gray(imagen)# La variable img imprime el en rgb2gray en una escala de RGB frios (morado, azul)
m1 = mm.ones((10,10)) / 25# Se define en el m1 (5,5) que tan opaca sera la intensidad de la imagen a menor cantidad mas desenfocada o opaca sera la misma.
img = convolve2d(img, m1, 'same')# El convolved mantiene la imagen en 2d y se le agrega los datos de la variable m1 que es la intensidad de la escala de rgb2gray y se le agrega un (same)que nos permite decir que es lo mismo. es decir la misma imagen pero con la escala.
img += 0.4 * img.std() * mm.random.standard_normal((img.shape))# En la variable img y junto a operador += se le multiplica un std, que es la desviacion estandar de array en numpy generando un cambio en la resolucion junto al shape donde buscamos las dimesiones del array que es lo que buscamos con la restauracion.
deconvolved_img = restoration.wiener(img, m1, 50)# Se le aplica la restauracion en una escala numerica junto a la varibale m1 y img donde deconvolved es quien anula la funcion convolve2d y se usa 
plt.imshow(deconvolved_img)# Junto a la biblioteca matplotlib se anula convolved y se genera la saturacion en la imagen y posterior se imprime.
plt.show()# Finalmente con la biblioteca matplotlib.pyplot se imprime la imagen modificada.
```
