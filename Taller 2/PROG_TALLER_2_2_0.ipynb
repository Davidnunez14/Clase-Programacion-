1. Escriba un algoritmo que dado un número entero, indique si este valor es par o impar.(recordar Todos los operadores aritméticos [+, -, *, /, %MOD])

**DIAGRAMA DE FLUJO:**

DIAG 1.png






```python
j=int(input("Ingrese un numero entero: ")) #Se define la variable que se va a utilizar en este caso "j" que sera un entero "int"
if j % 2==0: #Se usa un condicional "if" para dalre ordenes a la variable. Despues se utilizan los operadores aritmeticos en este caso % se usa para la division de enteros, y el "==" es usado para saber si un valor es igual a otro.
  print("El numero",j ,"es par") #Se deja un espacio TAB para dejar claro que la siguiente instruccion hace parte de la anterior. El print imprime los str condicionados por el if que sean pares
else: # else se utiliza para abarcar todos los valores que no fueron tomados por la condicion de IF (en este caso los impares)
  print("El numero",j ,"es impar") #El print imprime los str condicionados por el if que sean impares.
print("Fin...")
```

2. Escriba un algoritmo que pida un número y también un porcentaje entre 0 y 100 que
puede ser decimal, el algoritmo debe entregar el cálculo del porcentaje del primer
número, Si el usuario ingresa un dato fuera del rango (0 a 100), debe mostrar un
error y volver a pedir el porcentaje.

**DIAGRAMA DE FLUJO:**


Screenshot_14.png


```python
T=True #Evalua la expresion while, ejecutando los codigos que se encuentran en este
X=float(input("Escribe cualquier numero: ")) #X sera la variable que representa el numero que se va a ingresar para determinar su porcentaje el cual puede ser decimal
while(T): #Genera el bucle que se produciara cuando se ingrese un valor no valido para la variable "Y"
  Y=float(input("Escribe el porcentaje a calcular: ")) #Y sera la variable en la cual se inglesara el valor del porcentaje que se pienza calcular siendo este decimal
  if Y <0 or Y >100: #Se establece una condicion la cual consiste, en los valores que la variable "Y" del porcentaje arrojara un error  
    print("El porcentaje", Y ,"no es valido") #Este campo se imprime si el valor del porcentaje no es valido o esta duera del rango especificando en la casilla anterior
  else: #Sirbe para ejecutar el codigo que le sigue
    T=False #Si el valor de "Y" esta en el rango (0,100) no volvera a pedir este
R=Y*X/100 #Se efectua la operacion para sacar el porcentaje de la variable "X" con la "Y"
print("El", Y ,"% de", X ,"es de:", R) #Se imprime el resultado 
```

**EJERCICIO 3:**

---
3. Con base en el primer Ejercicio, si el número indicado es par, entregar todos los
números pares entre el 0 y el número.

**DIAGRAMA DE FLUJO:**

image.png





```python

j=int(input("Ingrese un numero entero: ")) #Variable que el usuario ingresa es un numero entero por el int

if j % 2==0:   #Se crea una condicion (si) para j (el numero del usuario) el cual cumple una formula matematica en la que el % se utiliza para dividir enteros, y el (==) para reconocer si es igual a un valor
   print("Su numero es par, y los numeros anteriores son:") # Este imprimirá si se cumple la condicion, esta dentro de la condicion para que siga unos pasos si cumple esa formula
   for i in range(2,j,2): #Se crea un ciclo y un rango el cual será de dos,hasta el numero del usuario y de dos en dos de manera consecutiva
    print(i, end=",") # imprimira el ciclo
else: #si no se cumple la primera condicion del if pasara a esta secuencia
    print("Su numero es inpar") #imprimira la condicion anterior,esta dentro de la condicion (else) para que siga unos pasos si cumple esa formula
```

**EJERCICIO 4:**

---
4. Escriba un algoritmo que dado un número entero, entregue la suma de todos los
números entre el 0 y el número dado.

**DIAGRAMA DE FLUJO:**

image.png


```python

n=int(input("Ingrese un numero entero: "))#Numero que ingresa el usuario
b=0   #variable para sumar los numeros

for i in range(1,n+1):  # ciclo que crea desde 0 hasta el numero del usuario
  if i ==1:
   print("",str(i),end="") #imprimira los numeros del ciclo hasta el numero del usuario,el "end" nos ayuda a ponerlo en linea
   b=b+i  # variable que guarda la suma de los numeros
  else:
   print(" + ",str(i),end="") #imprimira un mas y los numeros del ciclo hasta el numero del usuario,el "end" nos ayuda a ponerlo en linea
   b=b+i  # variable que guarda la suma de los numeros

print(end=" = ") #agrega un igual antes del resultado    
print(b) # resultado de la operacion
```
