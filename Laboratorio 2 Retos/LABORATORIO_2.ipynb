2. Escriba una función long_long() que tome una lista de palabras y devuelva la más larga.

```python
def longer_long(lista): #se define la funcion 'longer_long' con el argumento 'lista' con el c ual se ejecutaran las condiciones
  mas_larga="" #se defini una variable
  x=lista.split() #se define una segunda variable en donde 'split' dividira o separara la lista introducida en partes
  mas_larga = x[0] #se hace una igualacion entre las dos variables y se le define un valor a x
  for i in range(1, len(x)): #se crea un bucle el cual recorre los datos introducidos y se le aplica un rango y 'len' para que devuelva los datos
    if len(mas_larga)<len(x[i]): #se estable una condicion y se introducen 'len' para que compare y devuelva los valores 
      mas_larga=x[i] #se hace una igualacion y a 'x' se le coloca el valor definido en el bucle
  return mas_larga #se usa el 'return' para que retorne la igualacion
l=input('Introduzca una lista de palabras:') #se define la lista de la funcion mediante valores que se le piden al operador
print('La palabra mas larga es:',longer_long(l)) #se imprime  el resultado
```

4. Cree un pequeño programa que convierta números binarios en enteros. Ejemplo:

```python
def para_decimal(binario): #Se crea la funcion 'para_decimal' con el argumento 'binario' que es el que llevara las condiciones
    list = [] #La lista se crea para almacenar los valores que se anadiran 

    for j, i in enumerate(binario[::-1]): #El ciclo for se usa para empezar a operar los valores tal como se haria en una operacion a mano. Se dan dos variables 'j' e 'i'. La funcion de phyton 'enumerate' ayuda a separar los valores del numero binario. Y finalmente el'::-1' se usa para que empiece a tomar los valores de atras para adelante
        a = 2** j # Se define una anueva variable 'a' en la cual se va a elevar el numero de la variable j a la 2
        multiplicacion = int(i)*a #Se convierte en entero la variable 'i' para poderlo multiplicar por la variable creada 'a' y se denomina 'multplicacion' para usarla en el siguiente paso
        list.append(multiplicacion) #Se agregan los valores que dieron de los puntos anteriores a la lista con la funcion 'append'

    return sum(list) #Se usa el 'return' para que retorne el valor de la suma de la lista

x = input('Escibe un numero binario: ') #Se define el input para que el usuario ingrese el numero 
print('El resultado decimal es:',para_decimal(x)) #Dara el reslutado del numero binario en numero decimal 
```

6. Escribe una función es bisiesto () que determine si un año dado es un año bisiesto Un año bisiesto es divisible por 4, pero no por 100. También es divisible por 400.

```python
  año= int (input("Ingrese el año a evaluar: \n")) #se define la variable a la cual se ingresaran valores enteros 
def es_bisiesto(año): #se define la funcion y su argumento
   if año% 4==0 and año% 100!=0: #se establece una condicion 
    return "El año evaluado es bisiesto." #se retorna el valor de la condicion anterior
   elif año%100==0 and año%400==0: #se establece una condicion ejecutando los bloques de comando
        return "El año evaluado es bisiesto."#se retorna el valor de la segunda condicion
   else: #se coloca para controlar el flujo del programa y avarcar las condiciones que no se mensionan
    return "El año evaluado no es bisiesto." #se retorna el valor de las condiciones que abarca el 'else'

print(es_bisiesto(año))#se hace la invocacion de la funcion y se imprime
```

7. Escribe un programa que pida dos palabras y diga si riman o no. Si las tres últimas letras coinciden, tiene que decir que riman. Si sólo coinciden las dos últimas, tiene que decir que riman un poco y si no, que no riman.

EJEMPLO:

```python
palabra1 = input("Escriba una palabra ")# Se realiza una variable para la palabra 1
palabra2 = input("Escriba la segunda palabra: ")#Se realiza una variable para la palabra 2
 
tresultimas1 = palabra1[-3:] # Se realizan 4 variables las cuales va a revisar las 3 ultimas letras de las palabras.
tresultimas2 = palabra2[-3:]
tresultimas11 = palabra1[-2:]
tresultimas2 = palabra2[-2:]
 
if tresultimas1 == tresultimas2:# se crea una condicion la cual es que si las tres ultimas de primera palabra son iguales con las tres ultimas de la segunda palabra, las palabras rimas.
    print("Las palabras riman")
elif tresultimas11 == tresultimas2: # Si coinciden 2 letras de las 2 palabras dara como resultado rima un poco.
    print("Las palabras riman un poco")
else:         
    print("Las palabras no riman")#Si no es ninguna de las anteriores su resultado sera las palabras no riman.
```
