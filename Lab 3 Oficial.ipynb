```python
sec_asignada='AGTAGATGTAGTTAACTTTAATCTC' #cadena la cual va a buscar el codigo.
def busqueda(target,cadena): #se establece una funcion de nombre buscar y sus variables 'target y cadena'
  start=0 #se establece una variable de inicio para el comando
  end=len(sec_asignada) #se establece una variable destinada al largor de la cadena
  while (target): #se crea un bucle indefinido para que recorra todo el texto
    x=target[start:end] #se crea una variable que contenga las variables 'start y end'
    if x==cadena: #se establece una condiciovalor de  que compare el x con el de la cadena
      tupla=(start, end) #se crea una tupla que almacene los valores del comienzo y del final
      print('La cadena empieza en el caracter',start,'y finaliza en el',end) #se imprime para verificar que se cumple la condicion
      print(tupla) #se imprime el valor de la tupla      
      break #se rompe el ciclo cuando se cumple la condicion
    else: #se evaluan las condiciones no planteadas
      start+=1 #se suma un valor desde el comienzo asta que se cumpla la condicion planteada con el if
      end+=1 #se suma un valor al final asta que se cumpla la condicion de if
busqueda(secuencia_SAR,sec_asignada) #se invoca la funcion de busqueda que encontrara la secuencia del genoma que se nos asigno.
```
