```python
import json
import requests
from pprint import pprint

response = requests.get("https://fp.josedomingo.org/lmgs/u05/ej2.json")
pendientes=json.loads(response.text)

print("-------------------")

#EJERCICIO 1:
contador1=0
for Titulo in (pendientes):
  if Titulo["Titulo"]:
   contador1+=1
print("Hay",contador1,"Pruebas de idiomas") 

print("-------------------")

#EJERCICIO 2:

for Tiempo in (pendientes):
  if Tiempo["Horas"]>=2:
   print(Tiempo["Titulo"],"la duracion es de: ",end=" ")
   print(Tiempo["Horas"])

print("-------------------")

#EJERCICIO 3:

for nopresencial in (pendientes):   
  if nopresencial["TipoFormacion"]=="No Presencial": #En el archivo JSON el url correspondiente al idioma aleman esta compuesto de la siguiente manera "NoPresencial" por ende no se enseña
    print("URL de la informacion No Presencial son: ",nopresencial["URL"])

print("-------------------")

#EJERCICIO 4:
def ejercicio4profesores(jp,id):
    da = []
    for mm in jp:
        if mm["$id"]==id:
            da.append(mm["Titulo"])
            for profesor in mm["Profesorado"]:
                da.append(profesor["NombreCompleto"])
            return da
    return da

ejercicio4profesores(pendientes,str(input("Ingrese su ID de reconocimiento:")))

print("---------PUNTO 5----------")

#EJERCICIO 5:

profesores = []
jsonNombres = {}
jsonTitulos = {}
data = {}
data['names'] = []
data['titulos'] = []

for i in pendientes:
  titulos = i['Titulo']
  profesorado = i['Profesorado']
  jsonTitulos= json.dumps(titulos) #Se uso la libreria json para convertir el diccionario en un json.
  data['titulos'].append(titulos)
  print(jsonTitulos)

  for j in profesorado:
    nombres = j['NombreCompleto']
    jsonNombres = json.dumps(nombres)
    data['names'].append(nombres)
    print(jsonNombres)
    
    jsondatos = json.dumps(data) #Para convertir los titulos y profesores en un json.
print(jsondatos)


with open('profesores.json', 'w') as f:  #Para escribir en el archivo nuevo
  f.write(jsondatos)
```
